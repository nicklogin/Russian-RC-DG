```python
import re
import pandas as pd

from ast import literal_eval
```

```python
df = pd.read_csv("EGE_data/parsed_data.csv", sep=';', index_col="Unnamed: 0")
df["variants"] = df["variants"].apply(literal_eval)
df["right_answers"] = df["right_answers"].apply(literal_eval)
df["n_variants"] = df["variants"].apply(len)
df["n_right"] = df["right_answers"].apply(len)
```

```python
def question_subst_func(question: str) -> str:
    question = re.sub("\s+", " ", question).lower()
    if "какие из высказываний соответствуют" in question:
        return "Какое высказывание СООТВЕТСТВУЕТ тексту?"
    else:
        return "Какое высказывание НЕ СООТВЕТСТВУЕТ тексту?"
```

```python
df = df[df["n_variants"] != 0]
```

```python
df["question"] = df["question"].apply(question_subst_func)
```

```python
df_out = []

def alter_question(question: str) -> str:
    if question == "Какое высказывание СООТВЕТСТВУЕТ тексту?":
        return "Какое высказывание НЕ СООТВЕТСТВУЕТ тексту?"
    else:
        return "Какое высказывание СООТВЕТСТВУЕТ тексту?"

for index, row in df.iterrows():
    row_out = {"index": index, "reading_text": row["reading_text"]}

    question = row["question"]
    n_variants = row["n_variants"]
    n_right = row["n_right"]
    n_wrong = n_variants - n_right
    variants = row["variants"]
    right_ids = row["right_answers"]
    wrong_ids = [i for i in range(1, 6) if i not in right_ids]

    try:
        if n_right > n_wrong:
            row_out["question"] = alter_question(question)
            right_answer_id = wrong_ids[0] - 1
            row_out["right_answer"] = variants[right_answer_id]
            row_out["distractors"] = [variants[idx-1] for idx in right_ids]
        else:
            row_out["question"] = question
            right_answer_id = right_ids[0] - 1
            row_out["right_answer"] = variants[right_answer_id]
            row_out["distractors"] = [variants[idx-1] for idx in wrong_ids]
    except IndexError as e:
        print(wrong_ids, variants)
        raise e

    df_out.append(row_out)

df_out = pd.DataFrame(df_out).set_index("index")
```

```python
df_out
```

```python
def clear_sent_numbers(s: str) -> str:
    s = re.sub('\(?[0-9]+\)', ' ', s)
    s = re.sub('\s+', ' ', s)
    return s
```

```python
df_out["reading_text"] = df_out["reading_text"].apply(clear_sent_numbers)
df_out["right_answer"] = df_out["right_answer"].apply(clear_sent_numbers)
df_out["distractors"] = df_out["distractors"].apply(lambda x: [clear_sent_numbers(item) for item in x])
```

```python
import re, json

from typing import Union, Any
from math import ceil

import evaluate
import torch as tt
import pandas as pd

from transformers import AutoTokenizer, AutoModelForCausalLM
from transformers import PreTrainedModel, PreTrainedTokenizer
from transformers import T5Tokenizer, T5ForConditionalGeneration
from datasets import Dataset
from tqdm import tqdm_notebook
```

```python
# models:
tokenizer_rugpt3 = AutoTokenizer.from_pretrained("ai-forever/rugpt3small_based_on_gpt2")
model_rugpt3 = AutoModelForCausalLM.from_pretrained("RuGPT3-RuRACE/checkpoint-65760").to(tt.device("cuda:0"))
tokenizer_rugpt3.pad_token = tokenizer_rugpt3.eos_token

tokenizer_t5 = T5Tokenizer.from_pretrained("ai-forever/ruT5-base")
model_t5 = T5ForConditionalGeneration.from_pretrained("RuT5-RACE-tf-1/checkpoint-65760").to(tt.device("cuda:0"))


# metrics:
bleu4 = evaluate.load("bleu")
sbleu = evaluate.load("sacrebleu")
rouge = evaluate.load("rouge")
meteor = evaluate.load("meteor")
```

```python
def to_gpt3_format(example: dict[str, Union[str, list[str]]]) -> str:
  example["distractors"] = [option.strip() for option in example["distractors"] if option]
  right_answer = example["right_answer"].strip()

  outp = example["reading_text"].strip() + " ВОПРОС: " + example["question"].strip()

  outp += f" ПРАВИЛЬНЫЙ ОТВЕТ: {right_answer}"
  outp += "\nНЕПРАВИЛЬНЫЕ ВАРИАНТЫ ОТВЕТА:"

  inp = outp

  distractors = ''
  for option in example["distractors"]:
      if option != right_answer:
          #print(option)
          outp += f"\n  {option}"
          distractors += f"\n  {option}"

  return {"inp": inp, "outp_expected": outp, "distractors": distractors,"right_answer": right_answer}
```

```python
ege_gpt3 = ege_dataset.map(to_gpt3_format)
```

```python
MAX_OUTPUT_LENGTH = 69 # as learned from translated RACE train dataset

def cut_last_break(input_: list[str]) -> list[str]:
    output = [s[:s.rfind('\n')] for s in input_]
    return output

def parse_options(input_: list[str]) -> list[str]:
    output = [s.strip() for s in input_]
    output = [set(option.strip() for option in s.split('\n')) for s in output]
    output = [sorted(list(s))[:3] for s in output]
    output = ['\n'.join(s) for s in output]
    return output

def get_metric_inputs(
    input_batch: list[str],
    model: PreTrainedModel, tokenizer: PreTrainedTokenizer
) -> list[str]:

    input_batch_ = tokenizer(input_batch, return_tensors="pt", padding=True)["input_ids"].to(tt.device("cuda:0"))

    input_length = input_batch_.shape[-1]
    
    with tt.no_grad():
        output_batch = model.generate(input_batch_, max_length=input_length + MAX_OUTPUT_LENGTH)
        output_batch = output_batch[:,input_length:]

    output = tokenizer.batch_decode(output_batch)
    del input_batch_
    del output_batch
    tt.cuda.empty_cache()

    output = cut_last_break(output)
    output = parse_options(output)

    return output

def compute_metrics(output: list[str], label_batch: list[str]) -> dict[str, Any]:
    metric_dict = {
        "bleu": bleu4.compute(predictions=output, references=[[label] for label in label_batch]),
        "sbleu": sbleu.compute(predictions=output, references=[[label] for label in label_batch]),
        "rouge": rouge.compute(predictions=output, references=label_batch),
        "meteor": meteor.compute(predictions=output, references=label_batch)
    }
    return metric_dict
```

```python
BATCH_SIZE = 1
N_STEPS = (len(ege_gpt3) // BATCH_SIZE) + 1

metrics_rugpt3 = []

for i in tqdm_notebook(range(N_STEPS), total=N_STEPS):
    slice = ege_gpt3[i*BATCH_SIZE:(i+1)*BATCH_SIZE]

    if slice["inp"]:
        output = get_metric_inputs(slice["inp"], model_rugpt3, tokenizer_rugpt3)

        distractors = parse_options(slice["distractors"])

        metric = compute_metrics(output, distractors)

        # код далее подходит только для батчей из одиночных примеров (BATCH_SIZE=1):
        metrics_rugpt3.append({
            "idx": slice["index"][0],
            "article": slice["reading_text"][0],
            "question": slice["question"][0],
            "right_answer": slice["right_answer"][0],
            "distractors": distractors[0],
            "output": output[0],

            "bleu": metric["bleu"]["bleu"],
            "sbleu": metric["sbleu"]["score"],
            "rouge1": metric["rouge"]["rouge1"],
            "rouge2": metric["rouge"]["rouge2"],
            "rougeL": metric["rouge"]["rougeL"],
            "rougeLsum": metric["rouge"]["rougeLsum"],
            "meteor": metric["meteor"]["meteor"],
        })
```

```python
def to_t5_format(example: dict[str, Union[str, list[str]]]) -> str:
  inp, label = '', ''
  example["distractors"] = [option for option in example["distractors"] if option]
  right_answer = example['right_answer']

  right_answer = right_answer.replace('"', "'")

  inp += example['reading_text'] + " " + f"ВОПРОС: {example['question']} "


  inp += f'ПРАВИЛЬНЫЙ ОТВЕТ: "{right_answer}".'
  inp += 'НЕПРАВИЛЬНЫЕ ВАРИАНТЫ ОТВЕТА: '

  options = example["distractors"]
  options = [
      option.replace('"', "'") for option in options if option != right_answer
  ]
  options = [
      f'"{option}"' for option in options
  ]
  label = "; ".join(options)
    
  return {"inp": inp, "right_answer": right_answer, "distractors": label}
```

```python
ege_t5 = ege_dataset.map(to_t5_format)
```

```python
def get_metric_inputs_seq2seq(
    input_batch: list[str], #label_batch: list[str],
    model: PreTrainedModel, tokenizer: PreTrainedTokenizer
) -> list[str]:
    input_batch_ = tokenizer(input_batch, return_tensors="pt", padding=True)["input_ids"].to(tt.device("cuda:0"))
    # label_batch_ = tokenizer(label_batch, return_tensors="pt", padding=True)["input_ids"]

    # output_length = label_batch_.shape[-1]

    with tt.no_grad():
        output_batch = model.generate(input_batch_, max_length=MAX_LEN)

    output = [
        sent.replace("<pad>", " ").replace("</s>", " ").strip() for sent in tokenizer.batch_decode(output_batch) # 
    ]
    
    del input_batch_
    del output_batch
    # del label_batch_
    tt.cuda.empty_cache()

    return output

def compute_metrics(output: list[str], label_batch: list[str]) -> dict:
    metric_dict = {
        "bleu": bleu4.compute(predictions=output, references=[[label] for label in label_batch]),
        "sbleu": sbleu.compute(predictions=output, references=[[label] for label in label_batch]),
        "rouge": rouge.compute(predictions=output, references=label_batch),
        "meteor": meteor.compute(predictions=output, references=label_batch)
    }
    return metric_dict
```

```python
MAX_LEN = 81 # as learned from translated RACE train dataset
BATCH_SIZE = 1
N_STEPS = (len(ege_t5) // BATCH_SIZE) + 1

metrics_t5 = []

for i in tqdm_notebook(range(N_STEPS), total=N_STEPS):
    slice = ege_t5[i*BATCH_SIZE:(i+1)*BATCH_SIZE]

    if slice["inp"]:
        output = get_metric_inputs_seq2seq(slice["inp"], model_t5, tokenizer_t5)

        distractors = [item.replace('\n', '').replace('  ',' ').replace('  ',' ').strip() for item in slice["distractors"]]

        metric = compute_metrics(output, distractors)

        # код далее подходит только для батчей из одиночных примеров (BATCH_SIZE=1):
        
        metrics_t5.append({
            "article": slice["reading_text"][0],
            "right_answer": slice["right_answer"][0],
            "question": slice["question"][0],
            "distractors": distractors[0],
            "output": output[0],

            "bleu": metric["bleu"]["bleu"],
            "sbleu": metric["sbleu"]["score"],
            "rouge1": metric["rouge"]["rouge1"],
            "rouge2": metric["rouge"]["rouge2"],
            "rougeL": metric["rouge"]["rougeL"],
            "rougeLsum": metric["rouge"]["rougeLsum"],
            "meteor": metric["meteor"]["meteor"]
        })
```
